**4. Correlations among numeric variables**



```python
df_corr=df.corr(); df_corr #this command is used to find the pairwise correlation of all numeric variables in the dataframe. Non-numeric variables are ignored. By default, pandas generates pearson correlation coefficients. 
#correlation values range between -1 to +1. If the variables are negatively correlated, we get a value between -1 and 0.  If they are positively correlated, we get a value between 0 and +1. 
#A value of zero tells us that there is no relationship between the variables.
```

```python
sns.heatmap(df_corr,annot=True) #this command generates a color-encoded matrix of the correlation coefficients.
plt.show()
```

```python
sns.pairplot(df_corr,kind="reg", plot_kws={'line_kws':{'color':'red'}}, height=1.5, aspect=1) #This command visually represents the correlations between different pairs of variables in the dataframe. 
#We have customized the line color to appear red. 
#A positive slope indicates a postive relationship and a negative slope indicates a negative relationship. 
plt.show()
```

**Observations**: Sepal Width negatively correlates with all the other variables.

 *Strongest positive correlation* is observed between petal length and petal width. 
 
 There is a strong positive correlation between the following variables; sepal length and petal length, sepal length and petal width.

**5. Scatter plot for Petal width vs. Petal Length**

```python
#using seaborn to generate a Scatter plot for Petal width vs Petal Length
sns.scatterplot(data=df, x='petal.width', y='petal.length') #feeding in the variables
plt.xlabel('Petal Width') #to label x axis
plt.ylabel('Petal Length') #to label y axis
plt.title("Scatter plot for Petal width vs Petal Length") #inserting a title for the plot
plt.grid() #inserting grid lines. 
plt.show()
```

```python
#we can also generate a scatterplot with a best fit line describing the linear relationship between Petal width and Petal length
sns.lmplot(data=df, x='petal.width', y='petal.length') 
plt.xlabel('Petal Width') #customizing x axis label
plt.ylabel('Petal Length') #customizing y axis label
plt.title("Scatter plot (with best fit line)\n" +r"for Petal width vs Petal Length") #inserting a title and customizing the long title to appear in multiple lines
plt.grid() #inserting grid lines. 
plt.show()
```

**Observations**: In addition to the correlation coefficients, the above generated scatterplot illustrates a strong positive linear relationship between *Petal width* and *Petal length*. 

This ofcourse does not prove that one causes the other. 

---




**6. Multi-dimensional plot depicting target feature (*Variety*) as the color (3rd) dimension.**



```python
#using seaborn to generate a multidimensional scatterplot
sns.scatterplot(x='petal.width', y='petal.length', hue='variety', data=df) #feeding in the variables
plt.xlabel('Petal Width') #customizing x axis label
plt.ylabel('Petal Length') #customizing y axis label
plt.title('Multi-dimensional scatterplot\n' + r"with 'Variety' as the color dimension") #inserting a title and customizing the long title to appear in multiple lines
plt.grid() #inserting grid lines
plt.show()
```

```python
#we can also generate a multi-dimensional plot with a best fit line describing the linear relationship between Petal width and Petal length for each variety. 
sns.lmplot(data=df, x='petal.width', y='petal.length', hue='variety') 
plt.xlabel('Petal Width') #customizing x axis label
plt.ylabel('Petal Length') #customizing y axis label
plt.title('Multi-dimensional plot (with best fit line)\n' + r'for each Variety (Petal width vs Petal Length)') #inserting title and customizing it to appear in multiple lines
plt.grid() #inserting grid lines. 
plt.show()
```

**Observations**: We can see a positive linear relationship between petal length and petal width for all three varieties - Setosa, Versicolor and Virginica. 

Virginica is the variety with comparatively longer petal length and petal width. Setosa is the variety with comparatively shorter petal length and petal width. 
Hence, we can conclude that Setosas have relatively smaller sized flowers, Versicolors have relatively medium sized flowers and Vriginicas have relatively bigger sized flowers. 

The above multi-dimensional scatterplot illustrates a strong positive linear relationship between petal width and petal length for Versicolor variety and a weak positive linear relationship between these two variables for Setosa and Virginica varieties. 
This can be further confirmed by looking at the category (variety) wise correlation coefficients.

**1. Correlation between Petal length and Petal width for each variety**

```python
df.groupby('variety')[['petal.length','petal.width']].corr().iloc[0::2][['petal.width']] 
#correlation by category. using iloc to generate a smaller matrix without repeating the same value multiple times
```

**Observations**: As mentioned earlier, these correlation values support our conclusion that there is a strong positive correlation between petal width and petal length for the Versicolor variety, while there is a weak correlation between petal length and petal width for Setosa and Virginica varieties.

**2. Category (Variety) wise comparision of quartiles and outliers**



```python
# A loop to generate descriptive statistics for each variety
for x in df['variety'].unique():
  print(x)
  print(df[df['variety'] == x].describe())
```

```python
fig, axes = plt.subplots(2, 2, figsize=(18, 10)) 

fig.suptitle('Comparison of varieties') #title for the plot

sns.boxplot(ax=axes[0, 0], data=df, x='variety', y='sepal.length') #command for the upper left subplot
sns.boxplot(ax=axes[0, 1], data=df, x='variety', y='sepal.width') #comand for the upper right subplot
sns.boxplot(ax=axes[1, 0], data=df, x='variety', y='petal.length') #command for the lower left subplot
sns.boxplot(ax=axes[1, 1], data=df, x='variety', y='petal.width') #command for the lower right subplot
plt.show()
```

**Observations**: Virginica has relatively bigger flowers with longer sepal length, petal length and petal width. 

Setosa stands out as a category with relatively smaller petal length, petal width and sepal length, but has relatively wider sepals.

Setosa has a comparatively lower standard deviation in petal length and petal width.

**3.Correlation for each variety with pairplot and heatmap**



---



```python
sns.pairplot(df, hue = 'variety') #pair plot command in seaborn. This simple command generates the entire plot - built on two basic figures, the histogram and scatterplot.
#default pairplot only plots numerical columns. Hence, using variety (categorical variable) for coloring.
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5)) 

fig.suptitle('Heatmap with correlation of individual varieties') #title for the plot


sns.heatmap(ax=axes[0], data=df[df['variety']=='Virginica'].corr(),annot=True) #command for 1st subplot


sns.heatmap(ax=axes[1], data=df[df['variety']=='Setosa'].corr(),annot=True) #command for 2nd subplot


sns.heatmap(ax=axes[2], data=df[df['variety']=='Versicolor'].corr(),annot=True) #command for 3rd subplot
plt.show()
```

**Observations**: The histogram on the diagonal shows us the distribution of each of the variables (for each category), while the scatterplots show the relationship between two pairs of variables. 

All four variables (for each category) appears to be normally distributed. 

There is a strong correlation between most pairs of variables for Versicolor and Virginica. 
Setosa is the category with weak correlations among the variables. 

Linear relationship between petal width and petal lenth is the strongest. Even for this pair of variables, Setosa has a weak linear relationship. 

Heatmap shown above is a numerical and visual representation of the correlation between different pair of variables for each variety.
